```python
import face_recognition
from PIL import Image,ImageFilter,ImageDraw,ImageFont
```

```python
jk = face_recognition.load_image_file("/content/IMG_9209 (1).JPG","RGB")
print(jk.shape)
```

```python
l= face_recognition.face_locations(jk,model="cnn") #hog quick and by default
l
top = l[0][0]
right=l[0][1]
bottom = l[0][2]
left = l[0][3]

jk_face = jk[top:bottom , left:right]
print(jk_face.shape)
jk_face_img = Image.fromarray(jk_face)

font = ImageFont.load_default()
photo_img = Image.fromarray(jk)
draw = ImageDraw.Draw(photo_img)
draw.rectangle((left,top,right,bottom),
    outline = (0,0,255), width=5)
draw.text((left ,bottom -50),"nidhi",font=font, fill = (255,0,0)) 
photo_img
```

```python
import face_recognition
from PIL import Image,ImageFilter,ImageDraw,ImageFont

bts_group = face_recognition.load_image_file("/content/download.jpg","RGB")
print(bts_group.shape)
bts_pil_img = Image.fromarray(bts_group)

fnt = ImageFont.load_default()
fl = face_recognition.face_locations(bts_group) #hog quick and by default
draw = ImageDraw.Draw(bts_pil_img)

face_count = len(fl)
print("there are {} faces \n".format(face_count))

for i in range (face_count):
  top,right,bottom,left = fl[i]
  print("the top {} top: {} bottom:{} left:{} right:{}".format(face_count,top,bottom,left,right))
  draw.rectangle((left,top,right,bottom),
    outline = (0,0,255), width=5)
  draw.text((left ,bottom -50),str(i),font=fnt, fill = (255,0,0)) 

bts_pil_img
```
